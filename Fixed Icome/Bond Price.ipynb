```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define bond characteristics
face_value = 1.671  # Face value of the bond
coupon_rate = 0.052  # Annual coupon rate (5%)
maturity = 5  # Maturity in years
ytm_values = np.linspace(0.02, 0.10, 9)  # Range of yield to maturity (YTM) values
```

```python
# Calculate coupon payments
coupon_payment = face_value * coupon_rate

# Calculate bond prices for different YTM values
bond_prices = []
for ytm in ytm_values:
    discounted_cashflows = [coupon_payment / (1 + ytm)**t for t in range(1, maturity + 1)]
    final_cashflow = (face_value + coupon_payment) / (1 + ytm)**maturity
    bond_price = sum(discounted_cashflows) + final_cashflow
    bond_prices.append(bond_price)
```

```python
# Create a visualization of bond prices
plt.figure(figsize=(10, 6))
plt.plot(ytm_values, bond_prices, marker='o', linestyle='-')
plt.title("Bond Prices vs. Yield to Maturity")
plt.xlabel("Yield to Maturity (YTM)")
plt.ylabel("Bond Price")
plt.grid(True)
plt.show()
```
