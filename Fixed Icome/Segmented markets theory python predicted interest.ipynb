```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Generate synthetic data for interest rates and factors
np.random.seed(0)
n_obs = 100
years = np.arange(1, n_obs + 1)
interest_rates = 0.03 + 0.02 * years + np.random.normal(0, 0.01, n_obs)
factor1 = np.random.normal(0, 1, n_obs)
factor2 = np.random.normal(0, 1, n_obs)
```

```python

# Create a DataFrame
data = pd.DataFrame({'Year': years, 'Interest_Rate': interest_rates, 'Factor1': factor1, 'Factor2': factor2})

# Split the data into training and testing sets
```

```python
# Split the data into training and testing sets
train_data = data.iloc[:80]
test_data = data.iloc[80:]
```

```python
# Define the independent variables (factors)
X_train = train_data[['Factor1', 'Factor2']]
X_test = test_data[['Factor1', 'Factor2']]

# Define the dependent variable (interest rate)
y_train = train_data['Interest_Rate']
y_test = test_data['Interest_Rate']
```

```python
# Create and train a linear regression model
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Predict interest rates for the test set
predicted_interest_rates = model.predict(X_test)
```

```python
predicted_interest_rates
```

```python
# Plot the actual and predicted interest rates
plt.figure(figsize=(10, 6))
plt.plot(test_data['Year'], test_data['Interest_Rate'], label='Actual Interest Rate', marker='o')
plt.plot(test_data['Year'], predicted_interest_rates, label='Predicted Interest Rate', marker='x')
plt.title('Actual vs. Predicted Interest Rates')
plt.xlabel('Year')
plt.ylabel('Interest Rate')
plt.legend()
plt.grid(True)
plt.show()
```
