```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
alpha = 0.8
beta = 0.06
rho = 0.8
```

```python
T = 100
time = np.arange(0, T, 1)
```

```python
exchange_rate = np.zeros(T)
interest_rate_differential = np.zeros(T)
```

```python
exchange_rate[0] = 1.0
interest_rate_differential[0]=0.0
```

```python
for t in range(1, T):
    shock =np.random.normal(0, 0.01)
    interest_rate_diffirential = beta * (exchange_rate[t-1] -1 ) + shock
    exchange_rate[t] = rho * exchange_rate[t-1] + alpha * interest_rate_differential[t]
```

```python
plt.figure(figsize=(12, 6))
plt.plot(time, exchange_rate, label='Exchange Rate ', color='blue')
plt.plot(time, interest_rate_differential, label='Interest Rate Differential', color='red')
plt.xlabel('Time')
plt.ylabel('Value')
plt.title('Dornbush Overshooting Model Simulations')
plt.grid(True)
plt.show()
```
