```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
datasets= ['EURUSD=X', 'GBPUSD=X', 'EURGBP=X']


for dataset in datasets:
    Ticker = yf.Ticker(dataset)
    data = Ticker.history(start='2023-09-01', end='2023-09-29')
    filename = f'{dataset}_data.csv'
    data.to_csv(filename)
    print(f"Download data for {dataset} and saved as {filename}")
```

```python
Ticker = 'EURUSD=X', 'GBPUSD=X', 'EURGBP=X'
start_date = '2023-09-01'
end_date = '2023-09-29'
data = yf.download(Ticker, start=start_date, end=end_date)
```

```python
data[['Close']].plot(figsize=(10,6));
plt.grid(True)
```

```python
EURGBP_data = data['Close']['EURGBP=X']
EURUSD_data = data['Close']['EURUSD=X']
GBPUSD_data = data['Close']['GBPUSD=X']
```

```python
Implied_EUR_USD_data = EURGBP_data * GBPUSD_data
```

```python
arbitrage_Opportunity = EURUSD_data - Implied_EUR_USD_data
```

```python
arbitrage_Opportunity
```

```python
arbitrage_Opportunity.plot(figsize=(12,6));
plt.xlabel('DATE')

plt.title('Arbitrage Opportunity')
```

```python
plt.figure(figsize=(12, 6))
plt.plot(EURUSD_data, label='EUR/USD')
plt.plot(Implied_EUR_USD_data, label='Implied EUR/USD Via GBP', linestyle='--')
plt.fill_between(arbitrage_Opportunity.index, 0, arbitrage_Opportunity, where=arbitrage_Opportunity>0, color='green', alpha=0.5, label='Arbitrage Opportunity')
plt.xlabel('Date')
plt.ylabel('Exchange Rate')
plt.title('Triangular Arbritrage Visualization')
plt.legend()
plt.grid(True)
plt.show()
```
