```python
def price_contract(in_dates, in_prices, out_dates,
out_prices, rate, storage_cost_rate, total_vol,
injection_withdrawal_cost_rate):
    volume = 0
    buy_cost = 0
    cash_in= 0
    last_date = min(min(in_dates), min(out_dates))

    # Ensure dates are in sequences
    all_dates = sorted(set(in_dates), min(out_dates))

    for i in range (len(all_dates)):
        # Processing code for each date
        start_date = all_dates[i]

        if start_date in in_dates:
            # Inject on these dates and sum up cash flows
           if volume <= total_vol - rate:
              volume += rate

              # Cost to purchases gas
              buy_cost += rate  *  in_prices[in_dates.index(start_date)]
              # Injection cost
              injection_cost = rate * injection_withdrawal_cost_rate
              buy_cost += injection_cost
              print('Injected gas on %s at a prices of %s'% (start_date, in_prices[in_dates.index(start_date)]))

           else:
             # We don't want to inject when rates is greater than total volume minus volume
            print('Injection is not possible on date %s as there is insufficient space in the storage facility' %start_date)

        elif start_date in out_dates:
            # Withdraw on these dates and sum cashflows
            if volume >= rate:
                volume -= rate
                cash_in += rate * out_prices[out_dates.index(start_date)]
                # Withdrawal cost
                withdrawal_cost = rate * injection_withdrawal_cost_rate
                cash_in -= withdrawal_cost
                print('Extraction is not possible on date %s  as there is insufficient volume gas stored'% start_date)
    store_cost = math.ceil ((max(out_dates)-min(in_dates)).days // 30) * storage_cost_rate
    return cash_in - store_cost -buy_cost
    # Example usage of price_contract()
    in_dates = [date(2022, 4, 1), date(2022, 2, 1), date(2022, 2, 21), date(2022,  4, 1)] # Injection dates
    in_prices = [20, 21, 20.5, 22] # Prices on the extraction
    out_dates = [date(2022, 1, 27), date(2022, 2, 15), date(2022, 3, 20), date(2022, 6, 1)] # Extrations dates
    out_prices = [23, 19, 21, 25] # Prices on the extractions days
    rate = 10000
    storage_cost_rate = 10000
    injection_withdrawal_cost_rate = 0.005
    max_storage_volume = 500000
    result = price_contract(in_dates, in_prices, out_dates, out_prices, rate, storage_cost_rate, max_storage_volume, injection_withdrawal_cost_rate)
    print()
    print(f"The value of the contract is: ${result}")












```
